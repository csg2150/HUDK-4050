# The Second Step
（Add crime data to the list）
 ******
### Search and find the data from https://ucr.fbi.gov/crime-in-the-u.s/2019/crime-in-the-u.s.-2019/topic-pages/tables/table-8/table-8.xls/view


```python
import pandas as pd
waitlist = pd.read_csv("waitlist.csv")
waitlist
```

# The Third Step
（Calculate safety index, quality index and diversity index）
******
### (1) safety index =  (5 * violentCrime / 4 + propertyCrime / 3) * 100 / population   
*violentCrime is more harmful than propertyCrime thus has a weight of 5; violentCrime has four types thus divided by 4; propertyCrime has three types thus divided by 3; the number of crimes increases with higher population thus divided by population*
### (2) quality index = (completion rate - admission rate + meanSAT/2400) / 3   
*admission rate is negative related to quality thus minus admission rate; both completion rate and admission rate are between 0 and 1; divide SAT average score by the sum score to get a value located between 0 and 1; the data of earnings after graduation are missed*  
### (3) diversity index = variance of undergraduate numbers of different races    
*if students of different races are at similar percentage, the variance of undergraduate should be small*

```python
waitlist['safetyIndex'] = (waitlist['violentCrime']/4*5 + waitlist ['PropertyCrime']/3) * 100 / waitlist['Population']
waitlist
```

```python
waitlist['qualityIndex'] = (waitlist ['C150_4'] - waitlist ['ADM_RATE'] + waitlist['SAT_AVG']/2400) / 3
waitlist
```

```python

sublist = waitlist.loc[:,'UGDS_WHITE':'UGDS_NRA']

waitlist['diversityIndex']=sublist.std(axis=1)

waitlist
```

# The Fourth Step
（Calculate safety rank, quality rank and diversity rank）
******
### (1) Sorting universities by the safety index from A to Z. The first one has top safety rank    

### (2) Sorting universities by the quality index from Z to A. The first one has top safety rank  

### (3) Sorting universities by the diversity index from A to Z. The first one has top safety rank    
  


```python
waitlist1 = waitlist.sort_values('safetyIndex')

waitlist1

lst = list(range(1,37))

waitlist1['safetyRank'] = lst

waitlist1
```

```python
waitlist2 = waitlist1.sort_values('qualityIndex', ascending=False)

lst = list(range(1,37))

waitlist2['qualityRank'] = lst

waitlist2
```

```python
waitlist3 = waitlist2.sort_values('diversityIndex')


lst = list(range(1,37))

waitlist3['diversityRank'] = lst

waitlist3
```

# The Fifth Step
（Calculate rank score and final rank）
******
### (1) Calculate the rank score by computing the average of safety rank, quality rank and diversity rank
*Assuming that the three aspects equally matter for Alex*   

### (2) Sorting universities by the rank score from A to Z and quality rank from A to Z.
*When more than one univerisities have the same rank score, rank them by quality index*

### (3) Sorting universities by the rank scores from A to Z. The first one has top final rank 

  


```python
waitlist3['rankAverage'] = (waitlist3['safetyRank'] + waitlist3['qualityRank']+ waitlist3['diversityRank']/3)
waitlist3
```

```python
waitlist4 = waitlist3.sort_values('qualityIndex')
waitlist5 = waitlist4.sort_values('rankAverage')


lst = list(range(1,37))

waitlist5['final Rank'] = lst

waitlist5
```
